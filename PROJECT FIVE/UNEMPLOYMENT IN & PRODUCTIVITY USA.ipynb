```python
df=pd.read_csv('Unemployment.csv')
df
```

```python
df.dtypes
```

```python
top_five = df.sort_values(['Jan','Feb','March','April', 'May', 'June'], ascending=False).head(5)
top_five
```

```python
top_five.plot(y=['Jan','Feb','March','April','May', 'June'],
              x='State', 
              figsize=[8, 5], 
              ylim=[4.5,7],
              kind='bar',
             title='Five states that had highest unemploymnent in first half of 2019')
plt.savefig('top.pdf')
```

```python
low_five = df.sort_values(['Jan','Feb','March','April', 'May', 'June'], ascending=False).tail(5)
low_five
```

```python
low_five.plot(y=['Jan','Feb','March','April','May', 'June'],
              x='State',
              figsize=[8, 5], 
              ylim=[2,3], 
              kind='bar',
             title='Five US states that had least unemployment rate in April, May and June')
plt.savefig('bottom-rate.pdf')
```
